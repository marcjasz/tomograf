```python
%matplotlib inline
from scanner import Scanner
from skimage import io, color, util
from matplotlib import pyplot as plt
import pydicom
import numpy as np
import math

# Setupowanie zmiennych:
detectors = 300
step = 0.008
angle_spread = math.pi
photo = 'img/Head.jpg'
```

```python
def normalize(num, bot, top):
    if num > top:
        return 1
    if num < bot:
        return 0
    return (num - bot)/(top - bot)


def normalize_photo(photo):
    bot, top = np.percentile(photo, (2, 90))
    print(bot, top)
    res = [[normalize(num, bot, top) for num in x] for x in photo]
    return res

img = None

if(photo[-4:] == '.dcm'):
    
else:
    img = io.imread(photo)

img = color.rgb2gray(img)

plt.imshow(img)
plt.show()
```

```python
print("XD")
scan = Scanner(img).to_square_img()
sinogram = scan.generate_sinogram(angle_spread, detectors, step)
plt.imshow(sinogram)
plt.show()
```

```python
inversed_photo = scan.inverse_radon_transform(sinogram, angle_spread, detectors, step)
inversed_photo = normalize_photo(inversed_photo)
plt.imshow(inversed_photo)
plt.show()
```
